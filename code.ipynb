Importing required libraries

```python
import pandas as pd
import tensorflow as tf
import sklearn 
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from tensorflow import keras
import xgboost
from xgboost import XGBRegressor
```

Reading csv files

```python
vdata = pd.read_csv("C:\\Users\\Hemanth M\\C\\Python\\Project\\test.csv")
vdata
```

To check holidays and add a new column to check if it was a holiday or not

```python
data["holiday"].unique() # checking all the unique days

train = data
train.holiday = train.holiday.fillna('nan')
isHoliday = []
train['holiday'] = train['holiday'].astype('string')

for col in train.holiday:
    if col == 'nan':
        isHoliday.append(0)
    else:
        isHoliday.append(1)

train['isHoliday'] = isHoliday
train.head()
```

```python
valid = vdata.copy()
valid.holiday = valid.holiday.fillna('nan')
isHoliday = []
valid['holiday'] = valid['holiday'].astype('string')

for col in valid.holiday:
    if col == 'nan':
        isHoliday.append(0)
    else:
        isHoliday.append(1)

valid['isHoliday'] = isHoliday
valid.head()
```

Ordinal encoding the weather_main column

```python
oe = OrdinalEncoder()
train['weather_main'] = oe.fit_transform(train[['weather_main']])
train.head()
```

```python
valid['weather_main'] = oe.transform(valid[['weather_main']])
valid.head()
```

Extracting time from time field and putting it in a new column

```python
train['date_time'] = train['date_time'].astype('string')
time = []

for field in train['date_time']:
    time.append(int(field[-5:-3]))
train['time'] = time

valid['date_time'] = valid['date_time'].astype('string')
time = []

for field in valid['date_time']:
    time.append(int(field[-5:-3]))
valid['time'] = time
train.head()
```

Creating model

```python
features = ['isHoliday', 'rain_1h', 'snow_1h', 'weather_main', 'time', 'clouds_all']
X = train[features]
y = train.traffic_volume

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)

model = keras.Sequential([
    layers.Dense(512, activation='relu', input_shape=[6]),
    layers.BatchNormalization(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dense(1),
])

model.compile(
    optimizer='adam',
    loss='mae',
)

early_stopping = tf.keras.callbacks.EarlyStopping(
    min_delta= 0.005,
    patience= 10, 
    restore_best_weights=True,
)

history = model.fit(
    train_X, train_y,
    validation_data=(val_X, val_y),
    batch_size=100,
    epochs = 20,
    callbacks = [early_stopping],
    verbose = 1
)
```

Predictions

```python
features = ['isHoliday', 'rain_1h', 'snow_1h', 'weather_main', 'time', 'clouds_all']
X_testing = valid[features]

pred = model.predict(X_testing)
pred.shape
pred = pred.reshape((19281))
```

Making it into a file

```python
output = pd.DataFrame({'id': valid.id,
                       'traffic_volume': pred})

output.to_csv('submission.csv', index = False)
```
